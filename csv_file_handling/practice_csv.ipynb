```python
#csv dictreader

import csv
fp=open(r'D:\Python\csv_file_handling\emp_table.csv','r')
res=csv.DictReader(fp)

for i in res:
    print(i)
```

```python
#csv dict writer

fp=open(r'D:\Python\csv_file_handling\demofile.csv','w',newline='')
res=csv.DictWriter(fp,['name','age','branch'])
res.writeheader()
res.writerow({'name':'abhinav','age':'24','branch':'ece'})
fp.close()
```

```python
#csv dict writer(append)

l=['name','age','branch']
fp=open(r'D:\Python\csv_file_handling\demofile.csv','a',newline='')
res=csv.DictWriter(fp,fieldnames=l)

res.writerow({'age':'33','name':'virat','branch':'sports'})
fp.close()
```
