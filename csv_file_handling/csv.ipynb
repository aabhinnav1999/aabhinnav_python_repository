```python
#10 wap name of the employee and manager's designation if manager works in dept 10 or 20.

import csv
fp=open(r'D:\Cloud Computing and Python\Python\csv_file_handling\emp_table.csv','r')
fr=csv.reader(fp)

emgr={}
for i in fr:
    emgr[i[1]]=i[3]
# print(emgr)

m={}
fp=open(r'D:\Cloud Computing and Python\Python\csv_file_handling\emp_table.csv','r')
fr=csv.reader(fp)
for i in fr:
    if i[-1] in ('10','20'):
        m[i[1]]=i[0]
# print(m)

res1={}
for i in emgr:
    for j in m:
        if emgr[i]==m[j]:
            # print(i,j)
            res1[i]=j
# print(res1)

fp=open(r'D:\Cloud Computing and Python\Python\csv_file_handling\emp_table.csv','r')
fr=csv.reader(fp)
res2={}
for i in fr:
    for j in res1:
        if i[1]==res1[j]:
            res2[res1[j]]=i[2]
# print(res2)

for i in res1:
    for j in res2:
        if res1[i]==j:
            print(i,res2[j])

```

```python
def test(st):
    l=st.split('-')
    l[0],l[-1]='19'+l[-1],l[0]
    if l[1]=='DEC':
        l[1]='12'
    elif l[1]=='NOV':
        l[1]='11'
    elif l[1]=='OCT':
        l[1]='10'
    elif l[1]=='SEP':
        l[1]='09'
    elif l[1]=='AUG':
        l[1]='08'
    elif l[1]=='JUL':
        l[1]='07'
    elif l[1]=='JUN':
        l[1]='06'
    elif l[1]=='MAY':
        l[1]='05'
    elif l[1]=='APR':
        l[1]='04'
    elif l[1]=='MAR':
        l[1]='03'
    elif l[1]=='FEB':
        l[1]='02'
    else:
        l[1]='01'
    st='-'.join(l)
    return st

import csv
fp1=open(r"D:\Python\csv_file_handling\emp_table.csv",'r')
fr1=csv.reader(fp1)

for i in fr1:
    if i[4]!='HIREDATE':
        print(i[1],test(i[4].upper()),'--------',i[4].upper())
        
```
