```python
#2

import csv

fopen=open(r"D:\Python\csv_file_handling\aws_costs.csv",'r')
fr=csv.reader(fopen)

res=[]
for i in fr:
    if i[1].startswith('EC2')==False and i[0].startswith('2022')==True:
        print(i)
        res+=[i]

print('-'*160)

for i in res:
    print(i)
    for j in range(0,len(i)):
        if i[j]=='' or i[j]=='0':
            i[j]='0.00'

print('-'*160)

for i in res:
    print(i)
```

```python
#3

import csv

fopen=open(r"D:\Python\csv_file_handling\aws_costs.csv",'r')
fr=csv.reader(fopen)

z=[]
for i in fr:
    print(i)
    for j in range(0,len(i)):
        if j==0:
            print(i[j][3:])
            z+=[i[j][3:]]
        else:
            z+=[i[j]]
    break

print(z)
    
fnew=open('aws_final.csv','w',newline='')
fw=csv.writer(fnew)
fw.writerow(z)
fnew.close()
```

```python
#4

fopen=open(r"D:\Python\csv_file_handling\aws_costs.csv",'r')
fr=csv.reader(fopen)

z=[]
for i in fr:
    if i[0].startswith('S')==True:
        print(i)
        z+=[i]
        
for i in z:
    print(i)
    for j in range(0,len(i)):
        if i[j]=='' or i[j]=='0':
            i[j]='0.00'

fnew=open('aws_final.csv','a',newline='')
fw=csv.writer(fnew)
fw.writerows(z)
fnew.close()
```
