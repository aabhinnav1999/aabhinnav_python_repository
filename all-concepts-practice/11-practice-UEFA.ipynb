```python
#2 storing in new csv file

import csv

fo=open(r"D:\Python\all-concepts-practice\raw-data-UEFA.csv",'r')
fr=csv.reader(fo)

z=[]
for i in fr:
    if i[0]!="Season":
        if '56'<=i[0][-2:]<='99':
            z+=[['19'+i[0][-2:],i[2].upper(),i[1][3:].upper()]]
        else:
            z+=[['20'+i[0][-2:],i[2].upper(),i[1][3:].upper()]]



f1=open(r'D:\Python\all-concepts-practice\UEFA-final-0.csv','a',newline='')
fw=csv.writer(f1)

fw.writerows(z)
f1.close()
```
