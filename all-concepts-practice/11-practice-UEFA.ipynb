```python
#12 fetch data from mysql

import pymysql

mydb=pymysql.connect(host="127.0.0.1",port=3306,user='root',password='root',database='aabhinnav')
cus=mydb.cursor()

f=0
w=0
r=0
def final(team):
    global f,w,r
    cus.execute("select * from uefa where winners=%s or runners_up=%s",[team,team])
    res=cus.fetchall()
    for all in res:
        f+=1
        print(all)
        if all[2]==team.upper():
            w+=1
        elif all[-2]==team.upper():
            r+=1
    if f==0:
        return 'once check team name'
    print('-'*30,"finals ==== {}".format(f),'titles === {}'.format(w),'runners_up === {}'.format(r),sep='\n')

final('BAYERN MUNICH')
```
