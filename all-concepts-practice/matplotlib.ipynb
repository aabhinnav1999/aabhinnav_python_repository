```python
#3

import pymysql,matplotlib.pyplot as plt

db=pymysql.connect(host="127.0.0.1",port=3306,user='root',password='root',database='aabhinnav')
cus=db.cursor()

cus.execute("select year(date),count(*) from wc group by year(date)")
res=cus.fetchall()

x=[]
y=[]
for i in res:
    print(i)
    x+=[int(i[0])]
    y+=[int(i[1])]

plt.bar(x,y,color='g')
plt.xlabel('worlcup in year')
plt.ylabel('no of centuries')
plt.title('bar graph')
plt.show()
```
