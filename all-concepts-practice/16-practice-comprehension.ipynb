```python
#1

#1 l=[return_value for var in coll]

st='hello'
l=[var for var in st]
print('1.',l)

#2 l=[return_value for var in coll if<cond>]

n='9100584972'
l=[int(var) for var in n if int(var)%2!=0]
print('2.',l)

#3 l=[return_value if<cond> else return_value for var in coll]

n='9100584972'
l=[int(var) if int(var)%2==0 else None for var in n]
print('3.',l)

#4 l=[retur_value for var in coll for var in coll]

coll=[['abhinav',25],['anudeep',20]]
l=[var for var in coll for var in var]
print('4.',l)

#5 l=[return_value for var in coll for var in coll if <cond>]

coll=[['abhinav',25],['anudeep',20]]
l=[var for var in coll for var in var if type(var)==str]
print('5.',l)

#6 l=[return_value if <cond> else return_value for var in coll for var in coll]

coll=[['john',40],['miller',35]]
l=[var if type(var)==str and len(var)==4 else None for var in coll for var in var]
print('6.',l)
```

```python
#2 extract palindromes

st='123212165504213554521131155122612321'

l={st[i:i+j] for i in range(0,len(st)) for j in range(0,len(st)) if st[i:i+j]==st[i:i+j][::-1] and len(st[i:i+j])>=3}
print(l)

```

```python
#3 set comprehension

st={'csk','mi','gt','lsg'}

l={i+' vs '+j for i in st for j in st if i!=j}
print(l)
```

```python
#4 tuple comprehension

st='hello'
t=(var for var in st)
print(t)  #returns generator object
print(tuple(t))

st='12345'
t=(int(var) for var in st if int(var)%2==0)
print(tuple(t))

coll=[['john',40],['miller',35]]
t=(var for var in coll for var in var)
print(tuple(t))
```

```python
#5 dict comprehension

l=[1,2,3,4,5]

d={var:var**2 for var in l}
print(d)

d={var:var**2 for var in l if var%2==0}
print(d)

d={var:var**3 if var%2==0 else None for var in l}
print(d)
```

```python
#6

l=[chr(var) for var in range(65,75)]
print(l)
```
