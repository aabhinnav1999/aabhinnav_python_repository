```python
#1

import csv

l=[ 
    {'student_name':'abhinav','student_age':24},
    {'student_name':'anudeep','student_age':20},
    {'student_name':'pavan kumar','student_age':30}
    ]

with open('1.csv','w',newline='') as fo:
    fw=csv.DictWriter(fo,fieldnames=['student_name','student_age'])
    fw.writeheader()
    fw.writerows(l)
```

```python
#2 adding a column

import csv

fo=open(r'D:\Python\all-concepts-practice\file-handling-csv\1.csv','r')
fr=csv.DictReader(fo)

mail=['abhinav@gmail.com','anudeep2003@gmail.com','pavan@gmail.com']
a=0
l=[]

for i in fr:
    # print(i)
    i['student_mail']=mail[a]
    l+=[i]
    a+=1
    
fo.close()

# print(l)

with open('1.csv','w',newline='') as fo:
    fw=csv.DictWriter(fo,fieldnames=['student_name','student_age','student_mail'])
    fw.writeheader()
    fw.writerows(l)
```

```python
#3

import csv,pickle

l=[
    ['abhinav', '24', 'abhinav@gmail.com'],
    ['anudeep', '20', 'anudeep2003@gmail.com'],
    ['pavan kumar', '30', 'pavan@gmail.com']
]

fo=open(r'D:\Python\all-concepts-practice\file-handling-csv\2.csv','a',newline='')
fw=csv.writer(fo)

for i in l:
    res=pickle.dumps(i[2])
    fw.writerow([i[0],i[1],res])
    # print(i[0])

fo.close()
```
