```python
#1 python to json (serialization or dumping)
#dumps

import json

d={'name':'aabhinnav','dept':'ece'}
# d='hello world'
print(d)
print(type(d))

res=json.dumps(d)
print(res)
print(type(res))
```

```python
#2 json to python (deserialization or loading)
#loads

print(res,type(res))
res=json.loads(res)
print(res,type(res))
```

```python
#3 dump

import json

d={'name':'abhinav','age':24,'dept':'ece'}
fo=open('json-1.txt','w')
json.dump(d,fo)
fo.close()
```

```python
#4 load

import json

fr=open('json-1.txt','r')
res=json.load(fr)
print(res,type(res))
fr.close()
```
