```python
#1 python to pickle (pickling)
#dumps

import pickle

d={'date':'19-03-2024','location':'telangana'}
print((d),type(d))

res=pickle.dumps(d)
print(res,type(res))
```

```python
#2 store in pkl file
#dump

import pickle

d={'name':'abhinav','age':24,'dept':'ece','dob':'19-04-1999'}
fo=open('pickle-1.pkl','wb')
pickle.dump(d,fo)
fo.close()
```

```python
#3 loads

import pickle

fo=open('pickle-1.pkl','rb')
r=fo.read()
print(pickle.loads(r))
```

```python
#4 load

import pickle

fo=open('pickle-1.pkl','rb')
print(pickle.load(fo))
```

```python
#5 csv to pkl

import pickle,csv

fo=open(r'D:\Python\all-concepts-practice\8-1.csv','r')
fr=csv.DictReader(fo)

d={}
a=0

for i in fr:
    a+=1
    # print(i)
    d[a]=i

# print(d)
    
fopen=open('pickle-2.pkl','wb')
pickle.dump(d,fopen)
fopen.close()
```

```python
#6

fopen=open('pickle-2.pkl','rb')
r=fopen.read()
print(r)

for i in r:
    print(i)
```
