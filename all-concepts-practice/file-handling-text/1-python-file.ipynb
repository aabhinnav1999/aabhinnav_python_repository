```python
#6 creating a pickle file and storing data

import pickle

fo=open('pickle-1.pickle','wb')

d={'this is pickle file':1,'just for testing':2}

# st='this is string'

res=pickle.dumps(d)

fo.write(res)

fo.close()

```

```python
#7 reading from pickle file (decryption)

fr=open('pickle-1.pickle','rb')
res=fr.read()

print(pickle.loads(res))

fr.close()
```
