```python
#2 specific exception

def phone(num):
    return int(num)

c=0

try:
    print(phone(int(input("enter your phone number:"))))

except ValueError:
    c+=1
    print('please enter valid number ---',c,'chance completed')

    try:
        print(phone(int(input("enter your phone number:"))))
    except ValueError:
        c+=1
        print('please enter valid number ---',c,'chances completed')
        print('last chance')
   
        try:
            print(phone(int(input("enter your phone number:"))))
        except ValueError:
            c+=1
            print('your 3 chances completed')
```

```python
#6 raise

def sample(a,b):
    if type(a)==type(b):
        c=a+b
        print(c)
    else:
        raise ValueError('MY ERROR')
    
sample(10,10.9)
```

```python
#7 creating custom execption

class PYSERROR(Exception):
    pass

def sample(a,b):
    if type(a)==type(b):
        c=a+b
        print(c)
    else:
        raise PYSERROR('my custom error')
    
sample(10,2.0)
```

```python
#9

class todayerror(Exception):

    def __init__(self,msg='error created by developer') -> None:
        self.msg=msg

    def __str__(self) -> str:
        return str(self.msg)
    
print('start')

a=10
if a==10:
    raise todayerror

print('stop')
```

```python
#10

def demo(a,b):
    if type(a)==type(b):
        return a+b
    else:
        return
    
print(demo(10,20.1))

assert demo(10,20)==30

assert demo(10,20)==40, 'must be 30'
```
