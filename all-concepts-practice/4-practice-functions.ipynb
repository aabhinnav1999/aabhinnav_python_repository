```python
#1 without args and without return

def demo():
    n=5
    start=1
    prod=1
    while start<=n:
        prod*=start
        start+=1
    print(prod)

demo()
```

```python
#2 with args and without return

def demo(n):
    start=1
    prod=1
    while start<=n:
        prod*=start
        start+=1
    print(prod)

demo(7)
```

```python
#3 without args and with return

def demo():
    n=6
    start=1
    prod=1
    while start<=n:
        prod*=start
        start+=1
    return prod
demo()
```

```python
#4 with args and with return

def demo(n):
    start=1
    prod=1
    while start<=n:
        prod*=start
        start+=1
    return prod
demo(4)
```

```python
#deep copy

import copy

ab=['apple',(1.1,2.2),{'name':'abhinav','age':24},'hi',[89,99]]
cd=copy.deepcopy(ab)

ab[0]='banana'
ab[-1][3:3]=['done']
ab[1]+=(3.3,)
ab[2]['name']='anudeep'
print(ab)
print(cd)
```

```python
#removes all occurances of a value

l=[0,11,44,22,33,44,55,66]

def uremove(coll,ele):
    if type(coll)!=list:
        return 1
    a=0
    for i in coll:
        if coll[a]==ele:
            del coll[a]
        a+=1
    return coll
uremove(l,44)
```
