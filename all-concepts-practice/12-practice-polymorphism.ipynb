```python
#1 operator overloading

class sample:
    date="02-03-2024"

    def __init__(self,num) -> None:
        self.num=num

    def __add__(self,other):
        return self.num+other.num
    
    def __sub__(self,other):
        return self.num-other.num
    
a=sample(10)
b=sample(20)

c=sample(45)
d=sample(36)

print(a+b)
print(c-d)
```

```python
#2 method overriding

class parent:
    x=100
    def __init__(self) -> None:
        print('parent class constructor method')
    
    def display(self,num):
        print(num*15)
        print("hello world from parent")

class child(parent):
    def __init__(self) -> None:
        print('child class constructor method')

    def display(self):
        print("hello world from child")

pobj1=parent()
cobj1=child()

pobj1.display(15)
child.display(cobj1)

# parent.display(cobj1,20)

print(parent.__dict__)
print(child.__dict__)
```

```python
#3 method overriding

class parent:
    x=19

    def __init__(self,m,n) -> None:
        self.berry=m
        self.cherry=n
        print('parent class')
        print(self.fetch())


    def fetch(self):
        return self.berry,self.cherry

class child(parent):
    y=27

    def __init__(self, m, n) -> None:
        self.berry=m
        self.cherry=n
        print('child class')
        print(self.fetch())

    def fetch(self):
        return self.berry,self.cherry

pobj1=parent('berry','cherry')
conj1=child('strawberry','blueberry')
        
```

```python
#4 method overriding

class sample:
    def __init__(self) -> None:
        print("first init")

    def __init__(self) -> None:
        print("second init method")

    @staticmethod
    def display():
        print("first display method")

    @staticmethod
    def display():
        print("second display method")


obj=sample()
obj.display()
        
```
