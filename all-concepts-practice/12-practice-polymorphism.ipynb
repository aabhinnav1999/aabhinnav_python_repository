```python
#1 operator overloading

class sample:
    date="02-03-2024"

    def __init__(self,num) -> None:
        self.num=num

    def __add__(self,other):
        return self.num+other.num
    
    def __sub__(self,other):
        return self.num-other.num
    
a=sample(10)
b=sample(20)

c=sample(45)
d=sample(36)

print(a+b)
print(c-d)
```
