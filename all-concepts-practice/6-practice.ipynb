```python
#2

a='global'                      #intializing global variable
print(a)

def outer():
    global a                     #global keyword
    a=a+' outer'                 #modifying global variable
    b='local'                    #intializing local variable
    print('outer function a:',a)
    print('outer function b:',b)

    def inner():
        global a                 #global keyword
        nonlocal b               #nonlocal keyword
        a=a+' inner'             #modifying global variable
        b=b+' variable'          #modifying local variable
        c='nonlocal variable'    #intializing nonlocal variable
        print('inner function a:',a)
        print('inner function b:',b)
        print('inner function c:',c)

    inner()
    print('outer function b:',b)
    
outer()

print(a)
```
