```python
#5 convert lc to up and vice versa

st="@1999HeLLO$$WoRlD"
start=0
res=''

while start<len(st):
    if 'A'<=st[start]<='Z':
        res+=chr(ord(st[start])+32)
    elif 'a'<=st[start]<='z':
        res+=chr(ord(st[start])-32)
    else:
        res+=st[start]
    start+=1

print(res)
```

```python
#8
'''Wap to count the number of even and odd numbers is given a list. 
i/p:[1,8,9,7,25,26,98,75,74,36,15]
o/p:{'even':5,'odd':6}'''

l=[1,8,9,7,25,26,98,75,74,36,15]
start=0
d={'even':0,'odd':0}

while start<len(l):
    if l[start]%2==0:
        d['even']+=1
    else:
        d['odd']+=1
    start+=1

print(d)
    
```

```python
#9
'''Input=[10,10,10,20,30,40,40,50,60,70,70,70]
o/p:[10,20,30,40,50,60,70]
Eliminate repeated values.
Without using built in functions, set, set compershive, membership operator'''

input=[10,10,10,20,30,40,40,50,60,70,70,70]
start=0
d={}

while start<len(input):
    end=0
    count=0
    while end<len(input):
        if input[start]==input[end]:
            count+=1
        end+=1
    d[input[start]]=count
    start+=1

print(d)
```

```python
#10 list sorting
#case 1

l=[4,0,2,3,1,6,5]
start=0
while start<len(l):
    end=0
    while end<len(l):
        if l[start]<l[end]:
            l[start],l[end]=l[end],l[start]
            print(l)
        end+=1
    start+=1
print('result -------',l)
```

```python
#11  case 1
#input={'google.com','facebook.in','python.edu','youtube.com'}
#op:{'6':['google','python'],'7':['youtube'],'8':['facebook']}

li={'google.com','facebook.in','python.edu','youtube.com','intellipaat.com'}
li=list(li)
start=0
res=''
t=[]

while start<len(li):
    z=li[start]
    end=0
    while end<len(z):
        if z[end]=='.':
            t+=[li[start][:end]]
        end+=1
    start+=1

print(t)

start=0
while start<len(t):
    end=0
    while end<len(t):
        if len(t[start])<len(t[end]):
            t[start],t[end]=t[end],t[start]
        end+=1
    start+=1

print('sorted by length --------',t)

start=0
d={}

while start<len(t):
    z=t[start]
    count=len(t[start])
    if count in d:
        d[count]+=[z]
    else:
        d[count]=[z]
    start+=1
print(d)
```

```python
#case 2

li={'oranges.co.in','google.com','india.gov.in','facebook.in','twitter.com','python.edu','youtube.com','intellipaat.com','banana.com','apples.in'}
li=list(li)
start=0
res=''
t=[]

#user defined split
while start<len(li):
    z=li[start]
    end=0
    count=0
    while end<len(z):
        if z[end]=='.':
            count+=1
            if count==1:
                t+=[li[start][:end]]
        end+=1
    start+=1

print(t)

#sorting by alphabetical order
start=0
while start<len(t):
    end=0
    while end<len(t):
        if t[start]<t[end]:
            t[start],t[end]=t[end],t[start]
        end+=1
    start+=1

print('sorted by alphabets --------',t)

#sorting by length
start=0
while start<len(t):
    end=0
    while end<len(t):
        if len(t[start])<=len(t[end]):
            t[start],t[end]=t[end],t[start]
        end+=1
    start+=1

print('sorted by length --------',t)

start=0
d={}

#adding to dicitonary
while start<len(t):
    z=t[start]
    count=len(t[start])
    if count in d:
        d[count]=[z]+d[count]
    else:
        d[count]=[z]
    start+=1
print(d)
```
