```python
#1 factorial

def fact(n,start=1,res=1):
    if start==n+1:
        return res
    res*=start
    return fact(n,start+1,res)
print(fact(5))


def prod(n,res=1):
    if n==0:
        print('=',res)
        return None
    if n!=1:
        print(n,'x ',end='')
    else:
        print(n,end=' ')
    res*=n
    prod(n-1,res)
prod(6)
```

```python
#2 sum of numbers in given range

def usum(end,start=1,res=0):
    if start==end+1:
        return res
    res+=start
    return usum(end,start+1,res)

usum(5)
```
