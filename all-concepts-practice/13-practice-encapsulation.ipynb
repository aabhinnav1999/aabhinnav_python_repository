```python
#1 public modifier

import datetime

class bank:
    bank_name='hdfc'

    @staticmethod
    def date(d=datetime.datetime.now()):
        return d.date()
    
print(bank.bank_name)
print(bank.date())
# print(bank.__dict__)
```

```python
#2 protected modifier

import datetime

class bank:
    _interest='12.5%'

    @staticmethod
    def _date(d=datetime.datetime.now()):
        return d.timestamp()

print(bank._interest)
print(bank._date())
# print(bank.__dict__)
```

```python
#3 private modifier

class bank:
    __password="012%021"

    @staticmethod
    def __customers():
        print("we have 150 customers")

    @classmethod
    def getter(cls):
        return cls.__password
    
class child(bank):
    pass
    
# print(bank.__dict__)

print(bank.getter())
print(bank.getter())
```

```python
#4 private access modifier in a class

class college:

	__name = 'pyspiders'
	__location = 'hyd'

	def __init__(self, name, age): 
		self.__name = name
		self.__age = age

	def __displayDetails(self):
		print("Name: ", self.__name)
		print("Age: ", self.__age)
	
	def accessPrivateFunction(self): 
		self.__displayDetails() 

	@classmethod
	def __classdetails(cls):
		print("college name :",cls.__name)
		print("college location :",cls.__location)

	@classmethod
	def fetch(cls):
		cls.__classdetails()

obj = college('abhinav',24)

print(obj.accessPrivateFunction())
print(college.fetch())
```
