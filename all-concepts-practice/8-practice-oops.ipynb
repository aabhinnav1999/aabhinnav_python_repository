```python
#7  classmethod

class emp:
    company_name="apple"

    @classmethod
    def fetch(cls):
        print(cls.company_name)

    @classmethod
    def mod(cls,new_name):
        cls.company_name=new_name

obj=emp()

emp.mod('amazon')
obj.mod('samsung')

emp.fetch()           #calling class method using class address
obj.fetch()           #calling class method using object address
```

```python
#8 objectmethod

class employee:
    company_name="apple"

    def __init__(self,name,age) -> None:
        self.name=name
        self.age=age

    def get(self):
        print(self.company_name,self.name)

emp1=employee('aabhinnav',24)
emp1.get()                            #calling object method using object address

emp2=employee('sai',25)
employee.get(emp2)                    #calling object method using class address(with object address)
    
```

```python
#9  staticmethod

class demo:

    def __init__(self,name,loc) -> None:
        self.name=name
        self.loc=loc

    @staticmethod
    def test(coll):
        res=''
        for i in coll:
            if 'a'<=i<='z':
                res+=chr(ord(i)-32)
            else:
                res+=i
        return res
        

obj1=demo('sai','hyd')

print(obj1.test('sai'))
print(demo.test('hai!'))
```

```python
#10

class computer:

    def __init__(self,name,ram,processor) -> None:
        self.name=name
        self.ram=ram
        self.processor=processor

    def call(self):
        for i in self.__dict__:
            print(i,':',self.__dict__[i])
        return '-----------------'
    
hp=computer('hp','16gb','i7')
dell=computer('dell','8gb','ryzen5')

print(dell.call())
print(hp.call())
```
