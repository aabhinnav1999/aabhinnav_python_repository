```python
# binary search

def binary_search(arr, target):
  """
  Performs an iterative binary search on a sorted array.

  Args:
      arr: The sorted array to search.
      target: The element to search for.

  Returns:
      The index of the target in the array if found, otherwise -1.
  """
  low = 0
  high = len(arr) - 1
  while low <= high:
    mid = (low + high) // 2
    if arr[mid] == target:
      return mid
    elif arr[mid] < target:
      low = mid + 1
    else:
      high = mid - 1
  return -1

# Example usage
arr = [2, 5, 8, 12, 16]
target = 12
index = binary_search(arr, target)

if index != -1:
  print(f"Target found at index: {index}")
else:
  print("Target not found")
```

```python
#case 2

st=(10,20,30,40,40,50,10,20,40,40)

def unique(s):
    res=[]
    for i in s:
        if i not in res:
            res+=[i]
    return res

def minoccur(st):
    uni=unique(st)
    min=len(st)
    ch=''
    for i in uni:
        count=0
        for j in st:
            if i==j:
                count+=1
        if min>count:
            min=count
            ch=i
    return ch,min

res=minoccur(st)
print(res)
```

```python
# user defined list reverse function

l=[4,3,5,7,2,1,6]
# print(l.reverse())
# print(l)

# print(id(l))

def urev(coll):
    if type(coll)!=list:
        return 'coll should be list'
    else:
        start=0
        m=len(coll)//2
        while start<m:
            coll[start],coll[-1-start]=coll[-1-start],coll[start]
            start+=1
        return coll
    
print(urev(l))

# print(id(l))
```
