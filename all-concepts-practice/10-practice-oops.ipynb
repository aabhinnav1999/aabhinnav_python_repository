```python
#1 inheritance

class apple:
    prod1='iphone'
    prod2='mac'
    def __init__(self,model):
        self.model=model

class new_apple(apple):
    prod3='airpods'


obj1=apple(15)

new_apple.prod2='apple vision pro'

# del new_apple.prod1
del apple.prod1
# del apple.prod1

print(apple.__dict__)
print(new_apple.__dict__)

obj2=new_apple('SE')

print(obj2.model)
```

```python
#2 constructor overriding

class A:
    x=10
    y=20
    def __init__(self,m,n) -> None:
        print("parent class constructor")
        self.m=m
        self.n=n

class B(A):
    def __init__(self, a, b) -> None:
        print("child class constructor")
        self.a=a
        self.b=b

pobj=A(2023,2024)
cobj=B('abhinav','anudeep')

print(A.__dict__)
print(pobj.__dict__)
print(B.__dict__)
print(cobj.__dict__)
```
