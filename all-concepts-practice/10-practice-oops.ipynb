```python
#1 inheritance

class apple:
    prod1='iphone'
    prod2='mac'
    def __init__(self,model):
        self.model=model

class new_apple(apple):
    prod3='airpods'


obj1=apple(15)

new_apple.prod2='apple vision pro'

# del new_apple.prod1
del apple.prod1
# del apple.prod1

print(apple.__dict__)
print(new_apple.__dict__)

obj2=new_apple('SE')

print(obj2.model)
```

```python
#2 constructor overriding

class A:
    x=10
    y=20
    def __init__(self,m,n) -> None:
        print("parent class constructor")
        self.m=m
        self.n=n

class B(A):
    def __init__(self, a, b) -> None:
        print("child class constructor")
        self.a=a
        self.b=b

pobj=A(2023,2024)
cobj=B('abhinav','anudeep')

print(A.__dict__)
print(pobj.__dict__)
print(B.__dict__)
print(cobj.__dict__)
```

```python
#3 constructor chaining

class sample:
    x=100
    def __init__(self,i,j) -> None:
        self.i=i
        self.j=j

class demo(sample):
    y=9
    def __init__(self,i,j,k,l) -> None:
        # super().__init__(i,j)
        # super(demo,self).__init__(i,j)
        sample.__init__(self,[i,i*10],[j,j*10])
        self.k=k
        self.l=l

print(sample.__dict__)
pobj=sample('hi','bye')
print(pobj.__dict__)
cobj=demo(200,300,19,99)
print(cobj.__dict__)
```

```python
#4 constructor chaining

class college:

    college_name='pyspiders'

    def __init__(self,name,age) -> None:
        # print("parent class object")
        self.name=name
        self.age=age

std1=college('abhinav',24)
print(std1.__dict__)

class clg1(college):
    def __init__(self, name, age,no,dept) -> None:
        super().__init__(name, age)
        # print("child class object")
        self.no=no
        self.dept=dept

class clg2(clg1):
    def __init__(self, name, age, no, dept,loc) -> None:
        super().__init__(name, age, no, dept)
        self.loc=loc


std1=clg2('abhinav',24,9100584972,'ece','hyd')
print(std1.__dict__)
```

```python
#5 method chaining

class sample:
    x=100
    def __init__(self,m,n) -> None:
        self.m=m
        self.n=n

    def display(self):
        print("first value is ---",self.m)
        print("second value is ---",self.n)

class demo(sample):
    y=200
    def __init__(self, m, n,k,l) -> None:
        super().__init__(m, n)
        self.k=k
        self.l=l
    def display(self):
        super().display()
        print("third value is ---",self.k)
        print("fourth value is ---",self.l)

obj2=demo(9,99,199,209)

# print(demo.display(obj2))
# print(demo.__dict__)

print(obj2.display())
```

```python
#6 method chaining

class first:

    def __init__(self,name,age) -> None:
        self.name=name
        self.age=age

    def func(self):
        start=0
        res=''
        while start<len(self.name):
            if "a"<=self.name[start]<='z':
                res+=chr(ord(self.name[start])-32)
            else:
                res+=self.name[start]
            start+=1
        return res
    
class second(first):

    def __init__(self, name,last_name,age) -> None:
        super().__init__(name, age)
        self.last_name=last_name

    def func(self):
        start=0
        res=''
        while start<len(self.last_name):
            if "a"<=self.last_name[start]<='z':
                res+=chr(ord(self.last_name[start])-32)
            else:
                res+=self.last_name[start]
            start+=1
        res=super().func()+' '+res

        return res
    

obj=first('abhinav',24)
print(obj.func())
obj2=second('pavan','kumar',24)
print(obj2.func())
```
