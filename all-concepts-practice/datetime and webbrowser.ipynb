```python
# copilot program

import datetime

# Define the start and end years
start_year = 2001
end_year = 2007

# Iterate through each year
for year in range(start_year, end_year + 1):
    # Iterate through each month
    for month in range(1, 13):
        # Create a date for the first day of the month
        first_day = datetime.date(year, month, 1)
        # Check if the first day is a Sunday (weekday() returns 6 for Sunday)
        if first_day.weekday() == 6:
            print(f"Sundays in {first_day.strftime('%B %Y')}:")
            print(first_day.strftime("%Y-%m-%d"))
            # Iterate through the remaining days of the month
            current_day = first_day + datetime.timedelta(days=7)
            while current_day.month == month:
                print(current_day.strftime("%Y-%m-%d"))
                current_day += datetime.timedelta(days=7)
            print()  # Add a newline for readability
```
