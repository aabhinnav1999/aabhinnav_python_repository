```python
#1

import datetime,webbrowser

z=5
a=1
b=99
while True:
    d=datetime.datetime.now()
    if d.minute == z and d.second in (9,19,24,34,39,49,54,59):
        if b==99:
            print(a,'----------','helloworld!','---------',d.minute,d.second)
            webbrowser.open('https://www.cricbuzz.com/')
            b=100
            a+=1
    elif d.minute ==z+1 :
        break
    else:
        b=99
```

```python
#2

import datetime,csv

def demo():
    z=0
    a=1
    b=99
    while True:
        d=datetime.datetime.now()
        if d.minute == z and d.second in (9,19,24,34,39,49,54,59):
            if b==99:
                fo=open('testing.csv','a',newline='')
                fa=csv.writer(fo)
                fa.writerow([a,'aabhinnav',str(d.hour)+':'+str(d.minute)+':'+str(d.second)])
                fo.close()
                b=100
                a+=1
        elif d.minute ==z+1 :
            break
        else:
            b=99

print(demo())
```
