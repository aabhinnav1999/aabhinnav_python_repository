```python
#2

from abc import ABC,abstractmethod

class sample(ABC):
    
    @abstractmethod
    def display():
        pass

    @abstractmethod
    def show():
        pass

class demo(sample):
    print('this is demo class')

    @staticmethod
    def display():
        pass

    @staticmethod
    def show():
        pass

obj=demo()
print(obj.display())
```
