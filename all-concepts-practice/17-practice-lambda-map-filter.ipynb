```python
#1 lambda

var=lambda a,b:a+b
print(var(10,20))
```

```python
#2

var=lambda a,b:a*b if type(a)==type(b) else None
print(var(10,20.0))
```

```python
#3 map

l=[0,1,2,3,4,5,6,7]

m=map(lambda a:a*2 if a%2==0 else None,l)
print(list(m))

m=map(lambda a:a*2 if a%2==0 else a*1,l)
print(list(m))
```

```python
#4 filter

l=[0,1,2,3,4,5,6,7]

f=filter(lambda a: a if a%2==0 else None,l)
print(list(f))
```

```python
#5 

l=[0,1,2,3,4,5,6,7]

def func(a):
    if a%2==0:
        return a*a
    else:
        pass

m=map(func,l)
print(list(m))

f=filter(func,l)
print(list(f))
```
