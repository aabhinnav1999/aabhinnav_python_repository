```python
#modifying a column in csv file

import csv

with open('2-csv-sample1.csv','r') as fp:
    fr=csv.reader(fp)
    l=[]
    for i in fr:
        l+=[i]

for i in l:
    if i[-1]=='hyd' and i[1]=='burger':
        print(i)
        i[-1]='mumbai'
        print(i)
print(l)    

# print(l)
# print(l[0][1])
# l[0][1]='fooditem'
# print(l[0][1])

with open('2-csv-sample1.csv','w',newline='') as fp:
    fw=csv.writer(fp)
    fw.writerows(l)

```
