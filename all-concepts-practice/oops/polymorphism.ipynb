```python
#1 method overriding

class sample:
    cs='class state'

    def __init__(self) -> None:
        pass

    def display(self):
        print('this is parent class object method')

class child(sample):

    def __init__(self) -> None:
        pass

    def display(self):
        print('this is child class object method')

obj1=sample()
print(obj1.display())

obj2=child()
print(obj2.display())
```

```python
#2 operator overloading

class sample:
    name='abhinav'

    def __init__(self,x) -> None:
        self.x=x

    def __add__(self,other):
        return self.x+other.x
    
    def __sub__(self,other):
        return self.x-other.x
    
a=sample(35)
b=sample(15)
print(a+b)
print(a-b)
```
