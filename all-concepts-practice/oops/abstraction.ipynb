```python
#3

from abc import ABC,abstractmethod

class sample(ABC):

    @abstractmethod
    def display():
        pass

    @abstractmethod
    def show():
        pass

class demo(sample):
    print('demo class')

    def display(self):
        a=10
        b=20
        return a*b
    
    def show(self):
        print('this show method')

obj=demo()
res=obj.display()
print(res)
```
