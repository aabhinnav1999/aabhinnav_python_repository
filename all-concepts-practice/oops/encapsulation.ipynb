```python
#1 public access modifier(public members)

class bank:
    name='HDFC'

    def __init__(self,branch) -> None:
        self.branch=branch

    def display(self):
        print('This is {} bank'.format(self.name))
        print('Branch location is {}'.format(self.branch))

obj=bank('KTDM')
print(obj.display())
```

```python
#2 protect access modifier(protected members)

class bank:
    name='HDFC'

    def __init__(self,location) -> None:
        self.location=location
        self._manager='abhinav'

obj=bank('hyd')
print(obj._manager)
```

```python
#3 private access modifier(private members)

class bank:
    name='HDFC'

    def __init__(self) -> None:
        self.location='KTDM'
        self._manager='abhinav'
        self.__bal='$ 1M'

    def __fetch_bal(self):
        return 'balance is {}'.format(self.__bal) 
    
    def secure(self):
        return self.__fetch_bal()

obj=bank()

# these will not work
# print(obj.__bal)
# print(obj.__fetch_bal())

print(obj.secure())

# we can't modify private members from ouside of class
obj.__bal='$ 1.5M'

print(obj.secure())
```
