```python
#2

class fruit:
    location='hyd'

    def __init__(self,name,cost) -> None:
        self.name=name
        self.cost=cost

orange=fruit('orange',150)
print(orange.location)
print(orange.__dict__)
```

```python
#3 objectmethod

class student:
    clg_name='pyspiders'

    def __init__(self,name,age) -> None:
        self.name=name
        self.age=age

    def fetch(self):
        for i in self.__dict__:
            print(i,':',self.__dict__[i])

abhinav=student('abhinav',25)
abhinav.fetch()
print('-'*50)
student.fetch(abhinav)
```

```python
#4 classmethod

class fruit:
    location='hyd'

    @classmethod
    def loc(cls,location):
        cls.location=location

obj1=fruit()
print(obj1.location)

fruit.loc('delhi')
print(obj1.location,'==> changed location name using class address')

obj1.loc('banglore')
print(obj1.location,'==> changed location name using object address')

```

```python
#5 single level inheritance

class parent:
    name='parent property'

    def __init__(self,x,y) -> None:
        self.x=x
        self.y=y

obj1=parent(44,55)

class child(parent):
    name='child property'

print(child.name)     
```

```python
#6 constructor chaining

class parent:
    name='parent property'

    def __init__(self,x,y) -> None:
        self.x=x
        self.y=y

obj1=parent(44,55)

class child(parent):
    name='child property'

    def __init__(self, x, y,a,b) -> None:
        super().__init__(x, y)
        self.a=a
        self.b=b

cobj1=child(44,55,10,20)
```

```python
#7 method chaining

class college:
    name='pyspiders'
    loc='hyd'
    area='dilsukhnagar'

    @classmethod
    def fetch(cls):
        print(cls.name)
        print(cls.loc)
        print(cls.area)

print(college.fetch())
print('-'*50)

class child_college_1(college):
    area='kphb'

    @classmethod
    def fetch(cls):
        return super().fetch()
    
print(child_college_1.fetch())
print('-'*50)

class child_college_2(college):
    area='jntu'

    @classmethod
    def fetch(cls):
        return super().fetch()
    
print(child_college_2.fetch())
print('-'*50)
```
