```python
# positional args

def demo1(a,b,c):
    print(a,b,c)
    print(a+b+c)

demo1(15,12,19)
    
```

```python
# key-word args

def demo2(x,y,z):
    print('x is {}'.format(x),'y is {}'.format(y),'z is {}'.format(z),sep='\n')

demo2(y=199,z=214,x=19)
```

```python
# postional and keyword

def demo3(a,b,c,d):
    print(a,b,c,d)

demo3(10,20,d=100,c=200)
```

```python
# default

def demo3(m=1,n=5):
    for i in range(m,n+1):
        print(i,'square is',i*i)

demo3(3,19)
```

```python
# variable length args 
# packing

def pack1(*args):
    print(args)
    print(*args)

pack1(1,14,5)

def pack2(**kwargs):
    print(kwargs)
    # print(*kwargs)

pack2(name='abhinav',age=24)
```

```python
# unpacking

def unpack1(k,l,m):
    print(k,l,m)
unpack1(*{'aa','bb','cc'})

def unpack2(a,b,c):
    print(a,b,c)
unpack2(*'xuv')

def unpack3(name,sports):
    print(name,sports)
unpack3(**{'name':'worldcup','sports':'football'})

def unpack3(name,age):
    print(name,age)
unpack3(**{'name':'abhinav','age':24})
```
