```python
# Python code to illustrate Sending mail with attachments 
# from your Gmail account 

# libraries to be imported 
import smtplib 
from email.mime.multipart import MIMEMultipart 
from email.mime.text import MIMEText 
from email.mime.base import MIMEBase 
from email import encoders 

fromaddr = "businessaabhinnav@gmail.com"
toaddr = "abhinavkvd19@gmail.com"

# instance of MIMEMultipart 
msg = MIMEMultipart() 

# storing the senders email address 
msg['From'] = fromaddr 

# storing the receivers email address 
msg['To'] = toaddr 

# storing the subject 
msg['Subject'] = "Subject of the Mail"

# string to store the body of the mail 
body = "Body_of_the_mail"

# attach the body with the msg instance 
msg.attach(MIMEText(body, 'plain')) 

# open the file to be sent 
filename = "conditional statements.pdf"
attachment = open(r"D:\Python\conditional statements.pdf", "rb") 

# instance of MIMEBase and named as p 
p = MIMEBase('application', 'octet-stream') 

# To change the payload into encoded form 
p.set_payload((attachment).read()) 

# encode into base64 
encoders.encode_base64(p) 

p.add_header('Content-Disposition', "attachment; filename= %s" % filename) 

# attach the instance 'p' to instance 'msg' 
msg.attach(p) 

# creates SMTP session 
s = smtplib.SMTP('smtp.gmail.com', 587) 

# start TLS for security 
s.starttls() 

# Authentication 
s.login(fromaddr, "erna pnss znlm ohgu") 

# Converts the Multipart msg into a string 
text = msg.as_string() 

# sending the mail 
s.sendmail(fromaddr, toaddr, text) 

# terminating the session 
s.quit() 
```

```python
#3 attach documents

import smtplib
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText
from email.mime.base import MIMEBase
from email import encoders

fromadd="businessaabhinnav@gmail.com"
toadd="abhinavkvd19@gmail.com"

code=MIMEMultipart()

code['from']=fromadd
code['to']=toadd

code['subject']="here is your document(s)".upper()

body="body of the email"

code.attach(MIMEText(body,'plain'))

filename = "while looping.pdf"
attachment = open(r"D:\Python\while looping.pdf", "rb") 

# instance of MIMEBase and named as p 
p = MIMEBase('application', 'octet-stream') 
  
# To change the payload into encoded form 
p.set_payload((attachment).read())

encoders.encode_base64(p) 
   
p.add_header('Content-Disposition', "attachment; filename= %s" % filename)

code.attach(p)

s=smtplib.SMTP('smtp.gmail.com', 587)
s.starttls()

s.login(fromadd, "erna pnss znlm ohgu") 

text=code.as_string()

s.sendmail(fromadd,toadd,text)
s.quit()

print('success!')
```
