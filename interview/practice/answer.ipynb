```python
#3

import csv

fo=open(r'D:\Python\interview\practice\input.csv','r')
fr=csv.reader(fo)

a=0
l=[]
for i in fr:
    if i[0]=='Timestamp':
        a+=1
    if a==2:
        # print(i)    
        l+=[i]

for i in l:
    i[-1],i[-2]=i[-2],i[-1]
    if i[0]=='Timestamp':
        i[2:2]=[' metric2']
        i[100:100]=[' metric5']
        # print(i)
    else:
        i[2:2]=['None']
        i[5:5]=['None']
        # print(i)

l2=[]
for i in l:
    if i[0]!='Timestamp':
        l2+=[i]

fo=open('1-output.csv','a',newline='')
fw=csv.writer(fo)
fw.writerows(l2)
fo.close()
```
