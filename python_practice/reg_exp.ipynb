```python
#1

st="pyspiders <span>9988776622</span> qspiders 456 jsp<p>5566442233</P> test <h1>7788996655</h1>"
z=[]
res=""
l=st.split()


for i in l:
    for j in i:
        if '0'<=j<='9':
            res+=j
        else:
            if len(res)==10 and res[0] in '9876':
                z+=[res]
            res=""
    
print(z)
```
