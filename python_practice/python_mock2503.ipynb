```python
#2 

st='because is a conjunction because joins two sentences so because is a conjunction'
s=''
l=[]
start=0
while start<len(st):
    if st[start]!=' ':
        s+=st[start]
    else:
        l+=[s]
        s=''
    start+=1
l+=[s]
#print(l)

res=[]
for i in l:
    if i not in res:
        res+=[i]
#print(res)

for i in range(0,len(res)-1):
    for j in range(0,len(res)-1):
        if res[j]>res[j+1]:
            res[j],res[j+1]=res[j+1],res[j]
print(res)
```

```python
#3 

l1=['I','You']
l2=['play','Love']
l3=['chamma chakka','koti kommachi']

s=''
res=[]

for i in l1:
    for j in l2:
        for k in l3:
            s=i+' '+j+' '+k
            res+=[s]
            s=''
print(res)
```
