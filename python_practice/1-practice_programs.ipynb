```python
# user defined pop function
# case 1
def upop(coll,ind=-1):
    if ind ==-1:
        res=coll[ind]
        del coll[ind]
        return res
    if ind<0:
        ind =len(coll)+(ind)
    if ind>=len(coll) or ind<0:
        return 'out of index'
    res=coll[ind]
    del coll[ind]
    return res
l=[10,20,30,40,50,60,70]
res=upop(l,2)
print(res)
print(l)
        
```

```python
#case2
def upop(coll,ind=-1):
    if ind<-len(coll) or ind>=len(coll):
        return 'index out of range'
    if ind==-1:
        res=coll[ind]
        del coll[ind]
        return res,coll
    if ind<-1:
        res=coll[ind]
        del coll[ind]
        return res,coll
    res=coll[ind]
    del coll[ind]
    return res,coll
result=upop([1,2,3,4,5],-1)
print(result)
```

```python
# join function

def ujoin(coll,sub):
    s=''
    for i in coll:
        s+=str(i)+sub
    return s[:-(len(sub))]
ujoin(['ant','apple','bat','dog','fox'],',')
```

```python
#split function

def usplit(coll,sub):
    l=[]
    s=''
    for i in coll:
        if i !=sub:
            s+=i
        else:
            if len(s)!=0:
                l+=[s]
                s=''
    if len(s)!=0:
        l+=[s]
    return l
usplit('how are you venu',' ')
```

```python
# user defined hcf function

def hcf(a,b):
    start=1
    fact=1
    while start<=b:
        if a%start==0 and b%start==0:
            fact=fact*start
        start+=1
    return fact
hcf(int(input('enter first value')),int(input('enter second value')))
```

```python
#longest substring without repeating characters.

st='pwwkew'
res=[]
res_ele_len=[]

for i in range(0,len(st)):
    for j in range(i,len(st)):
        if len(set(st[i:j+1]))==len(st[i:j+1]):
            res+=[st[i:j+1]]
            res_ele_len+=[len(st[i:j+1])]

res.sort(key=len)


for i in res:
    if len(i)==max(res_ele_len):
        print(i)
        break
```
