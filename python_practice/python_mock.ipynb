```python
#2
start=0
st='hello@world#'
count=0
while start<len(st):
    if not('a'<=st[start]<='z' or 'A'<=st[start]<='Z' or '0'<=st[start]<='9'):
        count+=1
    start+=1
print(count)
```

```python
#3
a=[10,20,30,40]
b=type(a[0])
start=1
flag=True
while start<len(a):
    if type(a[start])!=b:
        print('not a homogenous list')
        flag=False
        break
    start+=1
if flag==True:
    print('homogenous list')
```

```python
#5
i=int(input())
j=int(input())
while i<=j:
    n=i
    count=0
    start=1
    while start<=n:
        if n%start==0:
            count+=1
        start+=1
    if count == 2:
        print(n)
    i+=1
```

```python
#6
t=('hello','haii',{10:1,20:2,30:3},(1,2,3),{'a':1})
res=[]
for i in t:
    if type(i) in (str,tuple) and len(i)%2!=0:
        res+=[i[::-1]]
    else:
        res+=[i]
print(res)
```

```python
#7
l=[1,2,3,'hello',2,'python','django','python','pyspiders','Django']
start=0
res=[]
while start<len(l):
    if type(l[start])==str and l[start] not in res and len(l[start])%2==0:
        res+=[l[start]]
    elif l[start] not in res and type(l[start])!=str :
        res+=[l[start]]
    start+=1
print(res)
```

```python
#8
st='Roses are Red'
start=0
l=[]
res=''
while start < len(st):
    if st[start]==' ':
        l+=[res[::-1]]
        res=''
    else:
        res+=st[start]
    start+=1
l+=[res[::-1]]
start=0
d={}
while start<len(l):
    i=0
    count=0
    z=l[start]
    while i<len(z):
        if z[i] in ('A','E','I','O','U','a','e','i','o','u'):
            count+=1
        i+=1
    d[l[start]]=count
    start+=1
print(d)
```
