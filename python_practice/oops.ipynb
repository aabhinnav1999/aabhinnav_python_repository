```python
class university:
    principal='geetha'
    loc='hyd'
    c=[]
    def __init__(self,name,dept,sal):
        self.name=name
        self.dept=dept
        self.sal=sal
        self.count(self)

    @classmethod
    def count(cls,self):
        cls.c+=[self]

e1=university('abhinav','ece',100000)
e2=university('anudeep','cse',200000)

import csv
fp=open(r'D:\Cloud Computing and Python\Python\python practice\oops-csv.csv','w',newline='')
res=csv.writer(fp)
li=[]
lo=[]
for i in e1.__dict__:
    li+=[i]
res.writerow(li)
for i in e1.__dict__:
    lo+=[e1.__dict__[i]]
res.writerow(lo)
fp.close()

print(university.c)
```

```python
#encapsulation

class A:
    x=10
    _y=20
    __z=30
    
    @classmethod
    def publicm(cls):
        print('this is public method')

    @classmethod
    def _protctm(cls):
        print('this is protect method')

    @classmethod
    def __privatem(cls):
        print('this is private method')

print(A.__dict__)
print(A._A__z)
A.publicm()
A._protctm()
A._A__privatem()
```

```python
#polymorphism

class sample:
    def __init__(self,x,y):
        self.x=x
        self.y=y

    def __add__(self,other):
        return self.x+other.x,self.y+other.y

    def __mul__(self,other):
        return self.x*other.x,self.y*other.y


obj1=sample(11,100)
obj2=sample(22,200)


print(obj1+obj2)
print(obj1*obj2)
```
