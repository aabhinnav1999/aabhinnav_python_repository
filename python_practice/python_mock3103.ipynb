```python
#1

def series(i,j,l=[]):      # this is recursion function
    if i==j+1:
        return l
    l+=[i]
    return series(i+1,j)
v=series(1,6)


def fact(l):
    start=0
    while start<len(l):
        i=1
        prod=1
        z=l[start]
        while i<=z:
            prod*=i
            i+=1
        print(prod)
        start+=1
fact(v)
```

```python
#2

def tab(i,j,n=19):
    if i==j+1:
        return 
    print(n,'*',i,'=',n*i)
    tab(i+1,j,n)
tab(1,10)
```

```python
#3

def fun(st,i=0,res=''):
    if i==len(st):
        return res
    if ('A'<=st[i]<='Z') and i%2!=0:
        res=chr(ord(st[i])+32)+res
    else:
        res=st[i]+res
    return fun(st,i+1,res)
st='PYSPIDERS'
fun(st)
```

```python
#4

def usplit(st,res='',l=[]):
    i=0
    while i<len(st):
        if st[i]!=' ':
            res+=st[i]
        else:
            l+=[res]
            res=''
        i+=1
    l+=[res]
    return l
st='i love india i love my country'
l=usplit(st)    

def z(l,i=0,res=[]):      # this is recursion function
    if i==len(l):
        return res
    if l[i] not in res:
        res+=[l[i]]
    return z(l,i+1,res)
res=z(l)

def result(l,res,d={}):
    for i in res:
        count=0
        for j in l:
            if i ==j:
                count+=1
        d[i]=count
    return d
result(l,res)
```
