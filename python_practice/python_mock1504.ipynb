```python
#1

st='BangaramTinnavara'
res1=''
result=''

for i in st:
    if i not in res1:
        res1+=i

for i in res1:
    if 'a'<=i<='z':
        result+=chr(ord(i)-32)
    else:
        result+=i
print(result)
```

```python
#2

st='HappyBirthDay'
res=''

for i in st:
    if i not in ('a','e','i','o','u','A','E','I','O','U') and 'A'<=i<='Z':
        res+=chr(ord(i)+33)
    elif i not in ('a','e','i','o','u','A','E','I','O','U'):
        res+=chr(ord(i)+1)
    else:
        res+=chr(ord(i)-32)

print(res)

```

```python
#3

st='HaPpYbirThDaY'
res=''

for i in st:
    if 'a'<=i<='z' and i not in ('a','e','i','o','u'):
        res=chr(ord(i)-32)+res

    elif 'A'<=i<='Z' and i in ('A','E','I','O','U'):
        res=chr(ord(i)+32)+res
        
    else:
        res=i+res
print(res)    
```

```python
#4

st1='BangaramTinnavara'

st=''
for i in st1:
    if 'A'<=i<='Z':
        st+=chr(ord(i)+32)
    else:
        st+=i

res=''
for i in st:
    if i not in res:
        res+=i


l=[]
for i in res:
    count=0
    for j in st:
        if i==j:
            count+=1
    if count>1:
        l+=[i]

result=''
for i in st:
    if i in l:
        result+=chr(ord(i)-32)
    else:
        result+=i
print(result)
```

```python
#5

st1='HappyBirthday'
st=''

for i in st1:
    if 'A'<=i<='Z':
        st+=chr(ord(i)+32)
    else:
        st+=i

res=''
count_con=0
count_vow=0
for i in st:
    if i in ('a','e','i','o','u'):
        count_vow+=1
        res+=i+str(count_vow)
    else:
        count_con+=1
        res+=i+str(count_con)

print(res)


```

```python
#6

st='@superModel0'

count_up=0
count_lo=0
count_sp=0
count_nu=0

for i in st:
    if 'A'<=i<='Z':
        count_up+=1
    elif 'a'<=i<='z':
        count_lo+=1
    elif not(('A'<=i<='Z') or ('a'<=i<='z') or ('0'<=i<='9')):
        count_sp+=1
    else:
        count_nu+=1

if count_up>=1 and count_lo>=1 and count_sp>=1 and count_nu>=1:
    print('strong password')
else:
    print('not a strong password')
    
        
```
