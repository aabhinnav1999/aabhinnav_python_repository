```python
#1 general syntax

def outer(func):
    def inner(*args,**kwargs):
        print('start')
        res=func(*args,**kwargs)
        print('stop')
        return res
    return inner

@outer
def sample(a,b):
    return a+b
print(sample(50,100))
```

```python
#2

def outer(func):
    def inner(*args,**kwargs):
        print('starting')
        res=func(*args,**kwargs)
        for i in range(*args):
            print(i**2)
        print('ending')
        return res
    return inner


@outer
def useries(a,b):
    l=[]
    for i in range(a,b):
        l+=[i]
    return l
print(useries(1,11))
```

```python
#3

def display_info(func):

    def inner():
        print('executing',func.__name__,'function')
        func()
        print('finished execution')

    return inner

@display_info
def printer():
    print('Hello World!')

printer()
```

```python
# decorator to a class

def outer(cls):
    def inner(*args,**kwargs):
        print('1.before execution')
        obj=cls(*args,**kwargs)
        print(obj.__dict__)
        for i in obj.__dict__:
            if i=='name':
                obj.__dict__[i]=input('modify name here:')
        print('2.after execution')
        print(obj.__dict__)
        print('-'*100)
        return obj
    return inner


@outer
class decorator:
    def __init__(self,name,date):
        self.name=name
        self.date=date
        
obj1=decorator('abhinav','0406')
obj2=decorator('ram','0506')


```
