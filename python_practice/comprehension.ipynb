```python
#6

'''Write a program to generate all sentences where the subject is in ["I", "You"] 
and the verb is in ["Play", "Love"] and the object is in ["Hockey","Football"].
Output:[“I Play Hockey”, “I Love Hockey”,“I Play Football”, “I Love 
Football”,
“You Play Hockey”, “You Love Hockey”,“You Play Football”, “You Love 
Football”, ]'''

s=["I", "You"] 
v=["Play", "Love"]
o=["Hockey","Football"]

l=[" ".join((i,j,k)) for i in s for j in v for k in o ]
print(l)
```

```python
#13 longest substring(not repeating characters) in a string.

st='javaconceptoftheday'

z=[st[i:j+1] for i in range(0,len(st)) for j in range(0,len(st)) if len(st[i:j+1])==len(set(st[i:j+1])) and len(st[i:j+1])>=2]
z.sort(key=len)
print(z[-1])
```
