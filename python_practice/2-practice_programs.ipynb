```python
# finding nth occurence index value user define function.

l=[10,20,10,50,47,99,30,10,20]

def uindex(coll,val,occ=1):
    count=0
    start=0
    while start<len(coll):
        if coll[start]==val:
            count+=1
        if count==occ:
            return start
        start+=1
print(uindex(l,50))
```

```python
#variable scope

x='global'
print(x)

def outer():
    global x
    x='local'
    print('local variable is:',x)

    def inner():
        global x
        x='nonlocal'
        print('nonlocal variable is:',x)
    inner()

    print(x)

outer()

print(x)
```

```python
#factorial using recursion

def rfact(n,start=1,prod=1):
    if start==n+1:
        return prod
    else:
        prod*=start
    return rfact(n,start+1,prod)
rfact(4)
```

```python
#1 print least palindrome in the string.

def demo(coll):
    res=[]
    for i in range(0,len(coll)):
        for j in range(i,len(coll)):
            if coll[i:j+1]==coll[i:j+1][::-1] and len(coll[i:j+1])>2:
                res+=[coll[i:j+1]]
    res.sort(key=len)
    return res[0]

demo('hipophellomomgood')
```

```python
# recursion (bidirectional)

def sample(num,start=1):
    if start==num:
        return
    print(start)
    sample(num,start+1)
    print(start)
sample(5)
```

```python
#deletion

l=[10,20,3,0,11,80]

def delete(coll,ele):
    for i in range(0,len(coll)):
        if ele==coll[i]:
            del coll[i]
            break
delete(l,0)

print(l)
```
