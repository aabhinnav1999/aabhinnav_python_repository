```python
#1 python to pickle

import pickle
data={'python':'pl','javascript':'pl','sql':'database'}

newfile=open('file1.pickle','wb')
res=pickle.dumps(data)
print(res)
newfile.write(res)
newfile.close()
```

```python
#2 pickle to python

import pickle

fo=open('file1.pickle','rb')
res=fo.read()
print(type(res),res)
r=pickle.loads(res)
print(type(r),r)
fo.close()
```
