```python
#1 python to json

import json

d={'name':'abhinav','age':24,'skills':['python','sql']}

fo=open('file1.json','w')
fo.write(json.dumps(d))
fo.close()
```

```python
#2 json to python

import json

fo=open(r'D:\Python\parsing\file1.json','r')
res=fo.read()
print(type(res),res)
r=json.loads(res)
print(type(r),r)
fo.close()
```
