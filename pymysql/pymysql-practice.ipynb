```python
#copying content from one csv file to another csv file.

import csv

with open(r'D:\Python\csv_file_handling\Untitled spreadsheet.csv','r') as fp:
    res=csv.reader(fp)
    for i in res:
        f1=open(r'D:\Python\csv_file_handling\Untitled spreadsheet-2.csv','a',newline='')
        fr=csv.writer(f1)
        fr.writerow(i)
    f1.close()
```

```python
#modifying particular column in csv

import csv

f1=open(r'D:\Python\csv_file_handling\Untitled spreadsheet-2.csv','r')
res=csv.reader(f1)

# res=csv.DictReader(f1)
# print(res)

li=[]
for i in res:
    li+=[i]
# print(li)

for i in li:
    print(i[2])
    

for i in li:
    z=i[2].split('-')
    # print(z)
    z[0],z[-1]=z[-1],z[0]
    z='-'.join(z)
    i[2]=z

print(li)

# for i in li:
fo=open(r'D:\Python\csv_file_handling\Untitled spreadsheet-2.csv','w')
fw=csv.writer(fo)
fw.writerows(li)
fo.close()
```

```python
#insert into table (csv to mysql)

import pymysql

mydb=pymysql.connect(host='127.0.0.1',port=3306,user='root',password='root',database='aabhinnav')
cus=mydb.cursor()

import csv
f1=open(r'D:\Python\csv_file_handling\Untitled spreadsheet-2.csv','r')
res=csv.reader(f1)

for i in res:
    if i[0]!='Name':
        cus.execute("insert into csvtable values(%s,%s,%s,%s,%s,%s,%s)",i)
        mydb.commit()


```
