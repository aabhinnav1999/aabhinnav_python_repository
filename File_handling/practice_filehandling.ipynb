```python
#1

fo=open(r'D:\Python\File_handling\harrypotter.txt','r')
res=fo.readlines()

s=''
for i in res:
    # print(i)
    for j in i:
        if j in ('a','A'):
            s+='b'

        elif j in ('e','E'):
            s+='f'

        elif j in ('i','I'):
            s+='j'

        elif j in ('o','O'):
            s+='p'

        elif j in ('u','U'):
            s+='v'
        else:
            s+=j

    f1=open('1-harrypotter.txt','a')
    f1.write(s)
    f1.close()
    s=''

# print(s)
```

```python
#2

fo=open('2-harrypotter.txt','r')
res=fo.read().split()
result=[]

# print(res)

for i in res:
    if res.count(i)>=2:
        if i not in result:
            result+=[i]

result=' '.join(result)
f2=open('2-harrypotter.txt','a')
f2.write(result)
f2.close()
```

```python
#3

fo=open(r'D:\Python\File_handling\harrypotter.txt','r')
res=fo.readlines()
# print(res[0].split())

for i in range(0,len(res)):
    z=res[i].split()
    # print(z)
    z[0],z[-1]=z[-1][::-1],z[0][::-1]
    # print(z)
    z=' '.join(z)
    with open('3-harrypotter.txt','a') as f3:
        f3.write(z+'\n')
```

```python
#4

fo=open(r'D:\Python\File_handling\harrypotter.txt','r')
res=fo.readlines()
# print(res)

for i in res:
    z=i.upper().split()
    z.sort()
    z=' '.join(z)
    f4=open('4-harrypotter.txt','a')
    f4.write(z+'\n')
    f4.close()
```

```python
#5

fo=open(r'D:\Python\File_handling\harrypotter.txt','r')
res=fo.readlines()

for i in range(0,len(res)):
    f5=open('5-harrypotter.txt','a')
    f5.write(res[i][-2::-1]+'\n')
    f5.close()
```
