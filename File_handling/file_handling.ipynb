```python
#1. Write a program to sort a string without using any function. (Write the raw logic).

st='shadowqwertyuoplkjhgfdsazxcvbnm'
res=''
l=[]
for i in st:
    l+=[i]

for i in range(0,len(l)-1):
    for j in range(0,len(l)-1):
        if l[j]>l[j+1]:
            l[j],l[j+1]=l[j+1],l[j]


for i in l:
    res+=i
print(res)
```

```python
#3 Write a program that accepts a number and returns the list of numbers that the input number is divisible with

num=int(input())
l=[]
for i in range(1,num+1):
    if num%i==0:
        l+=[i]
print(l)
```

```python
''' 4. Write a program that a number and then accepts strings numbering the initial
 input number (e.g. accepts 5 strings if the initial input number is 5).
Then sort the strings in reverse order and display the result'''

num=int(input('enter number:'))
start=1
l=[]
while start<=num:
    st=input('enter string:')
    l+=[st]
    start+=1
l.sort()
print(l)
```

```python
#8.Write a program that Reads a file and searches for a given String, and replaces that string everywhere with another given string.

st='Musk'
rt='abhinav'
fp=open('1.txt','r')
fr=fp.read()
l=fr.split()
fp.close()

start=0
while start<len(l):
    if st ==l[start]:
        l[start]=rt
    start+=1

data=' '.join(l)
fp=open('2.txt','w')
fr=fp.writelines(data)
fp.close()
```
