```python
input_list = ['methodology', 'a', 'is', 'devops', '', 'java script', 'css and', 'html', 'of ', 'web consists', '', 'language', 'programming', 'a', 'python is ']

# initialize temporary variables
temp_str = ''
output_list = []

# iterate through the input list
for word in input_list:
    if word == '':
        # add the temporary string to the output list and reset the variable
        output_list.append(temp_str.strip())
        temp_str = ''
    else:
        # append the word to the temporary string
        temp_str += ' ' + word

# add the final sentence to the output list
output_list.append(temp_str.strip())

# print the output list
print(output_list)
```

```python
#1 answer

fp=open(r'D:\Python\mock_file_handling_0505\question.txt','r')
z=fp.read()
l=z.split('\n')

res=[]

s=''
for i in l:
    # print(i)
    for j in i:
        # print(j)
        if j=='a' or j=='A':
            s+=chr(ord(j)+1)
        elif j=='e' or j=='E':
            s+=chr(ord(j)+1)
        elif j=='i' or j=='I':
            s+=chr(ord(j)+1)
        elif j=='o' or j=='O':
            s+=chr(ord(j)+1)
        elif j=='u' or j=='U':
            s+=chr(ord(j)+1)
        else:
            s+=j
    res+=[s]
    s=''
print(len(res))

for i in res:
    z1=open(r"answer1.txt",'a')
    z1.write(i)
    z1.write('\n')
    z1.close()
```
